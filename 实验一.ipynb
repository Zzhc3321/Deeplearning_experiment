```python
import tensorflow as tf
from tensorflow.keras import layers 
from tensorflow import keras

# 画图
from pyecharts.charts import Bar, Grid, Line, Liquid, Page, Pie, Bar3D
from pyecharts import options as opts
from pyecharts.globals import ThemeType
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

# 数据计算
import numpy as np    

# 鸢尾花数据集
from sklearn import datasets     
```

```python
iris = datasets.load_iris()
y_train = iris.target
x_train = iris.data
```

```python
print("y_train.shape:",y_train.shape)
print("x_train.shape:",x_train.shape)

print("y_train[1]:",y_train[1])
print("x_train[1]:",x_train[1])

print(len(set(list(y_train))))
print(set(list(y_train)))
```

```python
#动态分配显存
gpus = tf.config.experimental.list_physical_devices(device_type='GPU')
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)
    
model = keras.Sequential(
[
    layers.Flatten(input_shape=[4]),        
    layers.Dense(3, activation='softmax')
])
```

```python
model.compile(optimizer='SGD',                     
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train, batch_size=20, epochs=500, validation_split=0.1)
```

```python
xaxis_ = [i for i in range(len(history.history['accuracy']))]
```

```python
line_accuracy_chart = (
        Line(init_opts=opts.InitOpts(page_title='accuracy', height='600px', width='100%'))
            .add_xaxis(xaxis_)
            .add_yaxis("training_acc",history.history['accuracy'])
            .add_yaxis("valivation_acc",history.history['val_accuracy'])
            .add_yaxis(
                    series_name="training_acc",
                    y_axis=history.history['accuracy'],
                )
            .add_yaxis(
                series_name="valivation_acc",
                y_axis=history.history['val_accuracy'],
            )
            .set_series_opts(label_opts=opts.LabelOpts(is_show=False, font_size=5),
                         linestyle_opts=opts.LineStyleOpts(width=1, curve=0.1),)
            .set_global_opts(
                title_opts=opts.TitleOpts(title="accuracy"),
                tooltip_opts=opts.TooltipOpts(trigger="axis"),
                toolbox_opts=opts.ToolboxOpts(is_show=True),
                xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
            )
        )
```

```python
line_accuracy_chart.render("res_charts/01_accuracy.html")
```
